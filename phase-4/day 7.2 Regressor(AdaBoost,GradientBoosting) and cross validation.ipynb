```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
x = diabetes.data
y = diabetes.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gb_regressor = GradientBoostingRegressor(n_estimators=100,learning_rate=0.1,max_depth=3,random_state=42)
gb_regressor.fit(x_train,y_train)
```

```python
y_pred = gb_regressor.predict(x_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print("MSE:",mean_squared_error(y_test,y_pred))
print("MAE:",mean_absolute_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
print("R2_score:",r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred,color='blue',marker='o',label='Actual Vs Predicted')
plt.plot([min(y_test),max(y_test)],[min(y_pred),max(y_pred)],color='black',linestyle='dashdot',lw=2,label='Perfect Prediction')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual Vs Predicted Value (Gradient Boosting Regressor)')
plt.legend()
plt.grid(True)
plt.show()
```

#  AdaBoostRegressor


```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
ad_regressor = AdaBoostRegressor(n_estimators=100,learning_rate=0.1,random_state=42)
ad_regressor.fit(x_train,y_train)
```

```python
y_pred1 = ad_regressor.predict(x_test)
```

```python
print("MSE:",mean_squared_error(y_test,y_pred1))
print("MAE:",mean_absolute_error(y_test,y_pred1))
print(np.sqrt(mean_squared_error(y_test,y_pred1)))
print("R2_score:",r2_score(y_test,y_pred1))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred1,color='green',marker='o',label='Actual Vs Predicted')
plt.plot([min(y_test),max(y_test)],[min(y_pred1),max(y_pred1)],color='red',linestyle='--',lw=2,label='Perfect Prediction')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual Vs Predicted Value (Ada Boost Regressor)')
plt.legend()
plt.grid(True)
plt.show()
```
